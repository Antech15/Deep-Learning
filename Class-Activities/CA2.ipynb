# Binary Step

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5, .01)
y = np.where(x < 0, 0, 1)
plt.plot(x, y)
plt.title("Binary Step")
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()

x = np.array([-10000, -1000, -100, -10, -5, 0, 5, 10, 100, 1000, 10000])
y = np.where(x < 0, 0, 1)
print(y)
```

# Sigmoid

```python
def sigmoid(x):
  return 1/(1 + np.exp(-x))
x = np.arange(-5, 5, 0.01)
plt.plot(x, sigmoid(x))
plt.title('Sigmoid Activation Function')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()

x = np.array([-10000, -1000, -100, -10, -5, 0, 5, 10, 100, 1000, 10000])
print(sigmoid(x))
```

# Tanh

```python
def tanh(x):
  return ((np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x)))

x = np.arange(-5, 5, 0.01)
plt.plot(x, tanh(x))
plt.title('tanh')
plt.ylabel('g(x)')
plt.grid()
plt.show()

x = np.array([-10000, -1000, -100, -10, -5, 0, 5, 10, 100, 1000, 10000])
print(tanh(x))
```

# ReLu

```python
x = np.arange(-5, 5, 0.01)
y = np.maximum(0, np.where(x < 0, 0, x))

plt.plot(x, y)
plt.title('ReLu')
plt.ylabel('g(x)')
plt.grid()
plt.show()

x = np.array([-10000, -1000, -100, -10, -5, 0, 5, 10, 100, 1000, 10000])
y = np.maximum(0, np.where(x < 0, 0, x))
print(y)
```
