## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt

from math import erf #To calculate the erf() required in Gaussian Error Linear Unit (GeLU)
```

## Gaussian Error Linear Unit (GeLU)

```python
def gelu(x):
  return 0.5 * x * (1 + erf(x/(np.sqrt(2))))

gelu2 = np.vectorize(gelu) #fixes error

x = np.arange(-5, 5, 0.01)
plt.plot(x, gelu2(x))
plt.title("GELU")
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()

x = np.array([-10000, -1000, -100, -10, -5, 0, 5, 10, 100, 1000, 10000])
print(gelu2(x))
```

## Softplus

```python
def softplus(x):
  return np.log(1 + np.exp(x))

x = np.arange(-5, 5, 0.01)
plt.plot(x, softplus(x))
plt.title("Softplus")
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()

x = np.array([-10000, -1000, -100, -10, -5, 0, 5, 10, 100, 1000, 10000])
print(softplus(x))
```

## Exponential linear unit (ELU)

```python
#Using 1 for alpha

def elu(x, alpha):
  return np.where(x > 0, x, alpha * (np.exp(x) - 1))

x = np.arange(-5, 5, 0.01)
plt.plot(x, elu(x, 1))
plt.title("ELU")
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()

x = np.array([-10000, -1000, -100, -10, -5, 0, 5, 10, 100, 1000, 10000])
print(elu(x, 1))
```

## Scaled exponential linear unit (SELU)

```python
#alpha
a = 1.6732
#lambda
l = 1.0507

def selu(x, alpha):
  return np.where(x < 0, l * (alpha * (np.exp(x) - 1)), l * x)

x = np.arange(-5, 5, 0.01)
plt.plot(x, selu(x, a))
plt.title("SELU")
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.show()

x = np.array([-10000, -1000, -100, -10, -5, 0, 5, 10, 100, 1000, 10000])
print(selu(x, a))
```
